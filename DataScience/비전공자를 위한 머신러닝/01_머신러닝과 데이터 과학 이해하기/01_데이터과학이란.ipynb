## 1. 데이터 과학 Data Science
### 데이터 과학이 이슈가 된 이유
- 4차 산업혁명, 머신러닝, 인공지능 등 시대적 요구로 인해 데이터 자신으로 아이디어, 혁신을 구현하는 방식이 각광받기 때문.

### 사례
- PayPal에서 사기 탐지 프로그램을 만드는 데에 딥러닝이 사용됨. 기존보다 더 효율적인 방식으로 사기 관리?가 가능해졌다.
- 경영자의 직관, 감으로 진행되었던 일들이 데이터를 기반으로 한 의사결정이 각광받기 시작했고, 외국에서는 CDO라는 오피서가 등장함.

### 데이터 과학의 정의
- **컴퓨터**를 활용해서 **데이터**를 분석하고 **현실의 문제**를 해결하는 작업

### 데이터 과학의 발전 방향
1. Descriptive Analytics : (What) 기술 분석, 기술 통계, 현상에 대한 설명
2. Diagnostic Analytics : (Why) 왜? 그 현상이 발생했는지 원인, 이유를 찾고 과거를 돌아보게되면서 전통적인 통계 분석이 사용되게 됨.
3. Predictive Analytics : (Future) 단순히 과거에 발생다는 것이 아니라 미래에 어떻게 되는지 **예측**하게 됨.
4. Prescriptive Analytics : (Action) 미래 예측에서 끝나는게 아니라 그래서 어떻게 행동해야 하는지에 대해서를 다룬다. 자율주행같이 예측에서 끝나지 않고 **어떻게 행동해야할 지**를 생각하게 됨

> 예측 단계에서는 그저 통계적인 지식만 있으면 됐지만, 예측에 맞는 액션에 들어가게 되면서 도메인에 대한 이해, 현장에 대한 이해가 필요하게 되었다.

### 데이터 과학의 특징과 목표
- 데이터 과학은 융합형 인재를 원한다.
  - 예측에서 액션으로 넘어가게 되면서 하나의 단일화된 학문이 아니라 여러 학문이 섞이게 된 복합적인 학문이 되었다.
- 데이터 과학의 목표는 무엇인가?
  1. 의사 결정 (Decision Making) : 데이터 기반으로 생각하고 결정하게 됨
  2. 수익화 (Monetization) : 더 나은 의사결정으로 더 많은 수익을 얻기 위함. 데이터 과학으로 비즈니스를 더 잘하게 하는 것임.
