# 머신러닝(Machin Learning)이란 무엇인가
- 머신러닝과 다른 분석 방법 비교

## 빅데이터 분석 vs 통계 분석 vs 머신러닝
- 데이터를 활용해 분석하고 의사결정에 활용한다는 것은 공통점이다.

### 빅데이터 분석
- 상관관계를 파악하는 것에 주력.
- 예시 ) 그로스 해킹(Growth Hacking) : 프로그래밍을 포함한 기술 분석을 통한 마케팅
- 예시 ) 머리 면도를 하는 사람이 판매량이 많더라 라는 것을 상관관계 분석을 통해 알게 되었다. 직원들에게 면도기를 지급.
- 예시 ) 까마귀가 날아간 날엔 배가 떨어지더라!

### 통계 분석
- 실제 세계를 이해하고 설명하고 해석하는 데에 중점을 준다.
- 세상을 이해하기 위해 모델을 만들고, 통계치, 수치, 가정을 통해 파악하게 된다.
- 적은 데이터를 가지고 많이 해왔기 때문에 가정이 중요하다.

### 머신러닝
- 예측과 패턴 분석에 중점을 준다.
- 통계와 다르게 모형의 성능, 정확도가 더 중요하게 된다.
- 데이터가 많이 필요하다.

## 머신러닝은 무엇인가?
- 컴퓨터가 데이터로부터 스스로 학습을 할 수 있도록 프로그래밍하는 과학적 활동
- 현실적 문제 -> 가설 -> 검증 -> 검증검증 (검증하는 과정)
